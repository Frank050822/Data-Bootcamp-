# Data Cleaning

```python
DB.drop(["JOB TERMINATION DATE"],axis=1,inplace=True)
DB.head()
```

##### There are too many Zip Code and County, so we just dropped the job termination date. Because it may cause overfitted problem as we reviewed and assumed there are too many different values for each columns. We will not analyze job termination date.

```python
DB.dropna(subset=["AGE","GENDER","POSITION","EDUCATION","WAGE","HOURS PER WEEK"],inplace=True)
DB.head()
```

```python
DB["JOB_START_DATE"] = pd.to_datetime(DB["JOB START DATE"])
DB["PROGRAM START DATE"] = pd.to_datetime(DB["PROGRAM START DATE"])
```

##### In the above step we convert the data type of Date

```python
DB["DayDiff"] = DB["JOB_START_DATE"] - DB["PROGRAM START DATE"]
DB.head()
```

```python
DB["DayDiff"] = DB['DayDiff'].dt.days
DB.head()
```

##### DayDiff is how many days for candidates to get a job after submitting application.

```python
DB['type'] = np.where(DB['HOURS PER WEEK']>=40, 'FT', 'PT')
DB
```

```python
DB.dtypes
```

# Data Filtering

```python
FT = DB.loc[DB["HOURS PER WEEK"]>=40.0,:]
FT["TotalWages"] = FT["HOURS PER WEEK"] * FT["WAGE"]
FT.head()
```

### 3.1.2  Full-Time TotalWages 

#### 3.1.2 a) Linear Regression

```python
FT.plot.scatter(y='TotalWages',x='AGE')
```

##### The relationship between age and Totalwages is random

```python
reg_PTWages = smf.ols('TotalWages ~ AGE + GENDER + EDUCATION + POSITION ',data=FT).fit()
print(reg_PTWages.summary())
```

##### The P-value of Gender is very high, so we just remove this variable

```python
reg_FTWages = smf.ols('TotalWages ~ AGE + EDUCATION + POSITION ',data=FT).fit()
print(reg_FTWages.summary())
```

##### Since p-value of several dummy variables is still high, we have run regression again and then dropped the highest p-value once at each time until all p-value is lower than 5%. We just keep this process and got FT1 table to let the rest variables in the regression become significant.

```python
FT1=FT.loc[(FT['EDUCATION']!='12')&(FT['EDUCATION']!='11')].loc[(FT['EDUCATION']!='HS Graduate')&
           (FT['POSITION']!='Construction')&(FT['POSITION']!='College')&(FT['POSITION']!='Social Work')]
FT1
```

```python
reg_FTWages = smf.ols('TotalWages ~ AGE + EDUCATION + POSITION ',data=FT1).fit()
print(reg_FTWages.summary())
```

##### Therefore, we can get that PTWages = 4.8653* AGE - b1 * Position** - b2 * Education** + 557.6308 
##### Positon** and Education** are categorical data, so we need to match it with different coefficient. 
##### The coefficient of post graduate has the highest value to let the total wages be highest.  

```python
from sklearn.linear_model import LinearRegression as reg
from patsy import dmatrices
```

```python
y, X = dmatrices("TotalWages ~ AGE + POSITION + EDUCATION",data = FT)
```

```python
reg().fit(X,y).score(X,y)
```

#### 3.2.1 b) K-Nearest Neighbors

```python
from sklearn import neighbors
```

```python
k_FTE_range = range(1,101) 

scoresFTE_list=[]

for k in k_FTE_range:
    knnFTE = neighbors.KNeighborsClassifier(n_neighbors=k) 
    knnFTE.fit(X_train2, y_train2)
    
    scoresFTE_list.append(knnFTE.score(X_test2, y_test2))
```

##### We aim to select optimal K value that yields the highest accuracy. 

```python
plt.plot(k_FTE_range,scoresFTE_list)
plt.xlabel('Value of K')
plt.ylabel('Accuracy score')
```

```python
Accuracy_highest_FTE = max(scoresFTE_list)
K_HighestAccuracy_FTE = k_FTE_range[scoresFTE_list.index(Accuracy_highest_FTE)]
print ("Optimal K is",K_HighestAccuracy_FTE, "with highest accuracy =",Accuracy_highest_FTE)
```

```python
from sklearn import neighbors
knnFTE = neighbors.KNeighborsClassifier(n_neighbors=21)
knnFTE.fit(X_train2,y_train2)
knnFTE.score(X_test2,y_test2)
```
